```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
#avoiding access denial
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT x.y; Win64; x64; rv:10.0) Gecko/20100101 Firefox/10.0 '}
```

```python
rawdata = []
```

```python
for n in range(2002, 2022):
    URL = "https://www.lotterysearch.org/results/" + str(n)
    page = requests.get(URL, headers = headers)
    soup = BeautifulSoup(page.content, "html.parser")
    table = soup.find("div", id = "normal").find_all("tr")
    for n in range(1, len(table)-2):
        rawdata.append(table[n].get_text(separator = " ").split())
```

```python
df = pd.DataFrame(rawdata)
```

```python
df.head()
```

```python
df.columns = ["Day", "Date", "Month", "Year", "1", "2", "3", "4", "5", "6", "Lucky", "Jackpot", "Prize", "purge", "Draw"]
```

```python
df = df.drop(columns = "purge")
```

```python
df.head()
```

```python
df["Prize"] = [int(n[1:].replace(",", "")) for n in df.Prize]
```
